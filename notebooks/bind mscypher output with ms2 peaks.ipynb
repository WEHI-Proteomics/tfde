```python
import pandas as pd
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
```

```python
mscypher_filename = "/home/ubuntu/UPS2_allion/MSCypher.txt"
mscypher_df = pd.read_table(mscypher_filename, sep="\t", dtype="str", engine='python')
feature_id=4
DB_NAME = "/home/ubuntu/UPS2_allion/UPS2_allion-features-1-1097.sqlite"
```

```python
subset_df = mscypher_df[["DiggerPepScore", "FeatureNum", "FragMZ", "FragInt", "FragError", "FragIonTypes", "FragPos", "FragCharge", "DeltaMassDa", "DeltaMassPPM"]].copy()
```

```python
subset_df.head()
```

```python
subset_df = subset_df.apply(pd.to_numeric, errors='ignore').sort_values(by=['DiggerPepScore'], ascending=False)
```

```python
subset_df[subset_df.FeatureNum==feature_id]
```

```python
db_conn = sqlite3.connect(DB_NAME)
ms2_peaks_df = pd.read_sql_query("select feature_id,peak_id,centroid_mz,intensity from ms2_peaks_within_window where feature_id={}".format(feature_id), db_conn)
db_conn.close()
```

```python
ms2_peaks_df.head()
```

```python
frag_mz = subset_df[subset_df.FeatureNum==feature_id].iloc[0]
frag_mz
```

```python
FragMZ = list(map(float, frag_mz.FragMZ.split(';')))
FragInt = list(map(float, frag_mz.FragInt.split(';')))
FragError = list(map(float, frag_mz.FragError.split(';')))
FragIonTypes = list(map(str, frag_mz.FragIonTypes.split(';')))
FragPos = list(map(int, frag_mz.FragPos.split(';')))
FragCharge = list(map(int, frag_mz.FragCharge.split(';')))
```

```python
msc_fragments_df = pd.DataFrame(list(zip(FragMZ, FragInt, FragError, FragIonTypes, FragPos, FragCharge)), columns=['FragMZ', 'FragInt', 'FragError', 'FragIonTypes', 'FragPos', 'FragCharge'])
```

```python
msc_fragments_df
```

```python
msc_fragments_df["FragMZ_round"] = msc_fragments_df.FragMZ.round(3)
```

```python
ms2_peaks_df["centroid_mz_round"] = ms2_peaks_df.centroid_mz.round(3)
```

```python
df = pd.merge(ms2_peaks_df, msc_fragments_df, how='left', left_on=['centroid_mz_round'], right_on=['FragMZ_round'])
```
