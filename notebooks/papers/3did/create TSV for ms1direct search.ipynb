```python
experiment_name = 'P3856'
run_name = 'P3856_YHE211_1_Slot1-1_1_5104'
EXPERIMENT_DIR = '/media/big-ssd/experiments/{}'.format(experiment_name)
FEATURES_3DID_DIR = '/media/data-4t-a/results-P3856-3did/minvi-3000-2021-09-11-10-49-01/features-3did'
FEATURES_3DID_FILE = '{}/exp-{}-run-{}-features-3did-dedup.feather'.format(FEATURES_3DID_DIR, experiment_name, run_name)
```

```python
# load the features detected by 3DID
features_3did_df = pd.read_feather(FEATURES_3DID_FILE)
```

```python
features_3did_df.columns
```

```python
features_3did_df['rt_df'] = features_3did_df.apply(lambda row: json.loads(row.rt_df), axis=1)
```

```python
len(pd.DataFrame(features_3did_df.iloc[0].rt_df).frame_id.unique())
```

```python
features_3did_df['nScans'] = features_3did_df.apply(lambda row: len(pd.DataFrame(row.rt_df).frame_id.unique()), axis=1)
```

```python
features_3did_subset_df = features_3did_df[['monoisotopic_mass','rt_apex','rt_lower','rt_upper','charge','feature_intensity','nScans','isotope_count']].copy()
features_3did_subset_df.rename(columns={'monoisotopic_mass':'massCalib','rt_apex':'rtApex','feature_intensity':'intensityApex','isotope_count':'nIsotopes'}, inplace=True)
```

```python
features_3did_subset_df.to_csv('{}/{}-3did.features.tsv'.format(expanduser('~'),run_name), sep='\t', index=False, header=True)
```
