```python
import numpy as np
import pandas as pd
import pickle
from scipy import signal
import matplotlib.pyplot as plt
import peakutils
import math
```

```python
l = []
l.append((1,2,3))
l.append((11,22,33))
l.append((10,20,30))
l.append((12,22,32))
l.append((13,23,33))
l.append((14,24,34))
df_1 = pd.DataFrame(l, columns=['a','b','c'])
```

```python
l = []
l.append((1,'one'))
l.append((11,'eleven.1'))
l.append((11,'eleven.2'))
l.append((10,'ten'))
l.append((12,'twelve'))
l.append((13,'thirteen'))
l.append((14,'fourteen'))
l.append((15,'fifteen'))
df_2 = pd.DataFrame(l, columns=['a','word'])
```

#### \'inner\' means the intersection of the left and right dfs

```python
pd.merge(df_1, df_2, how='inner', left_on='a', right_on='a')
```

#### \'outer\' means the union of the left and right dfs

```python
pd.merge(df_1, df_2, how='outer', left_on='a', right_on='a')
```

#### \'right\' means the right df is the reference, and every row is populated, even if the left df doesn't have a match (they are filled with NaN

```python
pd.merge(df_1, df_2, how='right', left_on='a', right_on='a')
```
