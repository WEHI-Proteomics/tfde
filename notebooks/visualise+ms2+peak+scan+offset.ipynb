```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
merged_feature_db_name = '/home/daryl/subset-data/subset-Hela_20A_20R_500.sqlite'
db_conn = sqlite3.connect(merged_feature_db_name)
ms2_peak_matches_df = pd.read_sql_query("select * from precursor_ms2_peak_matches", db_conn)
db_conn.close()
```

```python
ms2_peak_matches_df
```

```python
indexes_to_drop = ms2_peak_matches_df.ms2_peak_id==0
ms2_peak_matches_df.drop(ms2_peak_matches_df.index[indexes_to_drop], inplace=True)
```

```python
fig, ax = plt.subplots()
ax.plot(ms2_peak_matches_df.mz_centroid, ms2_peak_matches_df.scan_delta, marker='o', linestyle='', ms=5)
plt.legend(loc='best')
plt.xlabel('m/z')
plt.ylabel('scan offset')
plt.margins(0.06)
plt.title('Scan offset of matched ms2 peaks relative to their precursor')
plt.gca().invert_yaxis()
plt.show()
```
