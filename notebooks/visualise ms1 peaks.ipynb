```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
db_name = 'z:/wilding-mcbride/BSA_All_Ion.sqlite'
db_conn = sqlite3.connect(db_name)
peaks_df = pd.read_sql_query("select * from summed_frames where frame_id=6486", db_conn)
db_conn.close()
```

```python
peaks_df
```

```python
unallocated_points_df = peaks_df[peaks_df.peak_id==0]
allocated_points_df = peaks_df[peaks_df.peak_id>0]
allocated_points_df
```

```python
fig, ax = plt.subplots()
ax.plot(unallocated_points_df.mz, unallocated_points_df.scan, marker='o', linestyle='', ms=5, color='tab:gray')
ax.plot(allocated_points_df.mz, allocated_points_df.scan, marker='o', linestyle='', ms=5, color='tab:orange')
plt.legend(loc='best')
plt.xlabel('m/z')
plt.ylabel('scan')
plt.margins(0.06)
plt.title('Frame')
plt.gca().invert_yaxis()
plt.show()
```
