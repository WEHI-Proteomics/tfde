```python
target_metrics_that_failed = {}
for row in df.itertuples():
    if row.target_metrics is None:
        if 'target_metrics' in target_metrics_that_failed:
            target_metrics_that_failed['target_metrics'] += 1
        else:
            target_metrics_that_failed['target_metrics'] = 1        
    elif None in row.target_metrics.values():
        for k,v in row.target_metrics.items(): 
            if v is None:
                if k in target_metrics_that_failed:
                    target_metrics_that_failed[k] += 1
                else:
                    target_metrics_that_failed[k] = 1
```
