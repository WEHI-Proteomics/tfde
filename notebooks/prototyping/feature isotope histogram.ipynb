```python
import pandas as pd
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
RT_LIMIT_LOWER = 4340
RT_LIMIT_UPPER = 4580
```

```python
BASE_NAME = "/home/ubuntu/HeLa_20KInt-rt-{}-{}".format(RT_LIMIT_LOWER,RT_LIMIT_UPPER)
feature_list_filename = "{}/HeLa_20KInt-feature-list.csv".format(BASE_NAME)
feature_list_df = pd.read_csv(feature_list_filename)
```

```python
feature_list_df.head()
```

```python
f, ax1 = plt.subplots()
f.set_figheight(5)
f.set_figwidth(8)
plt.hist(feature_list_df.isotope_count, bins=np.arange(start=2, stop=8)-0.5, rwidth=0.6)
plt.legend(loc='best')
plt.xlabel('number of isotopes')
plt.ylabel('feature count')
plt.margins(0.06)
plt.suptitle('Number of features with at least two isotopes')
plt.title('{}'.format(feature_list_filename))
plt.show()
```

```python
f, ax1 = plt.subplots()
f.set_figheight(5)
f.set_figwidth(8)
plt.hist(feature_list_df.charge_state, bins=np.arange(start=2, stop=8)-0.5, rwidth=0.6)
plt.legend(loc='best')
plt.xlabel('charge state')
plt.ylabel('feature count')
plt.margins(0.06)
plt.suptitle('Number of features with at least 2+ charge')
plt.title('{}'.format(feature_list_filename))
plt.show()
```
