```python
TARGET_DECOY_MODEL_DIR = '/Users/darylwilding-mcbride/Downloads/experiments/dwm-test/target-decoy-models'
EXPERIMENT_METRICS_FILE_NAME = "{}/experiment-metrics-for-library-sequences.pkl".format(TARGET_DECOY_MODEL_DIR)
```

```python
experiment_metrics_for_library_sequences_df = pd.read_pickle(EXPERIMENT_METRICS_FILE_NAME)
```

```python
experiment_metrics_for_library_sequences_df.sample(5)
```

```python
# unpack the TFD-extracted attributes from each sequence
attributes = []
attribute_names = [key for key in sorted(experiment_metrics_for_library_sequences_df.iloc[0].attributes)]

for row in experiment_metrics_for_library_sequences_df.itertuples():
    # target metric valuesa
    d = row.attributes
    if (d is not None):
        # check whether any of the values are None
        all_not_none = True
        for k,v in d.items(): 
            if d[k] is None:
                all_not_none = False
                break
        # add the attributes to the collection
        if all_not_none == True:
            l = []
            l.append(row.sequence)
            l.append(row.charge)
            l.append(row.file_idx)
            l += [d[key] for key in sorted(d)]
            attributes.append(tuple(l))

columns = ['sequence','charge','file_idx']
columns += attribute_names

attributes_as_extracted_df = pd.DataFrame(attributes, columns=columns)
```

```python
attributes_as_extracted_df.isotope_idx_not_in_saturation.unique()
```

```python
sequences_with_attributes_df = pd.merge(attributes_as_extracted_df, experiment_metrics_for_library_sequences_df, how='inner', left_on=['sequence','charge','file_idx'], right_on=['sequence','charge','file_idx'])
```

```python
sequences_with_attributes_df.isotope_idx_not_in_saturation.unique()
```

```python
sequences_with_attributes_df.sample(5)
```

```python
number_of_peptides_affected_by_saturation = len(sequences_with_attributes_df[sequences_with_attributes_df.isotope_idx_not_in_saturation > 0])
number_of_peptides_with_all_isotopes_in_saturation = len(sequences_with_attributes_df[sequences_with_attributes_df.isotope_idx_not_in_saturation == -1])
total_number_of_peptides = len(sequences_with_attributes_df)
print("proportion of peptides affacted by saturation: {}%".format(round(number_of_peptides_affected_by_saturation / total_number_of_peptides * 100, 1)))
print("proportion of peptides with all isotopes in saturation: {}%".format(round(number_of_peptides_with_all_isotopes_in_saturation / total_number_of_peptides * 100, 2)))
```
