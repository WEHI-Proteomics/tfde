```python
import pandas as pd
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
EXPERIMENT_BASE_DIR = '/media/big-ssd/experiments'
EXPERIMENT_NAME = 'P3830_YUPS1'
RUN_NAME = 'P3830_YeastUPS1_01_Slot1-1_1_5066'
```

```python
RAW_DATABASE_NAME = '{}/{}/raw-databases/{}.d/analysis.tdf'.format(EXPERIMENT_BASE_DIR, EXPERIMENT_NAME, RUN_NAME)
```

```python
# frame types for PASEF mode
FRAME_TYPE_MS1 = 0
FRAME_TYPE_MS2 = 8
```

```python
db_conn = sqlite3.connect(RAW_DATABASE_NAME)
frame_intensity_df = pd.read_sql_query("select Id,SummedIntensities,Time from Frames where MsMsType == {} order by Time".format(FRAME_TYPE_MS1), db_conn)
db_conn.close()
```

```python
f, ax1 = plt.subplots()
f.set_figheight(5)
f.set_figwidth(15)
ax1.set_ylabel('total frame intensity')
ax1.set_xlabel('retention time (secs)')
plt.axvline(x=1500, linestyle=':', color='tab:purple')
plt.axvline(x=2500, linestyle=':', color='tab:cyan')
ax1.plot(frame_intensity_df.Time, frame_intensity_df.SummedIntensities, marker='', linestyle='-', linewidth=0.5, ms=2, color='tab:green', label='frame')
plt.margins(0.06)
plt.suptitle('Total Ion Chromatogram (TIC) for raw ms1 frames')
plt.title('{}'.format(RAW_DATABASE_NAME))
plt.show()
```
