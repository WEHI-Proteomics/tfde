```python
import pandas as pd
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
feature_list_filename = "/home/ubuntu/HeLa_20KInt/HeLa_20KInt-feature-list.csv"
feature_list_df = pd.read_csv(feature_list_filename)
```

```python
ecdf = sm.distributions.ECDF(feature_list_df.summed_intensity)
```

```python
f, ax1 = plt.subplots()
plt.plot(ecdf.x, ecdf.y)
f.set_figheight(5)
f.set_figwidth(10)
plt.xscale('log')
plt.xlabel('log(intensity)')
plt.ylabel('ECDF')
plt.margins(0.02)
plt.suptitle("Feature Intensity Empirical Cumulative Distribution Function (ECDF)")
plt.title('{}'.format(feature_list_filename))
plt.show()
```
